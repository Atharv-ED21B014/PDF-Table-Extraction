```python
import pdfplumber
from PIL import Image
import pytesseract
from img2table.document import Image as Img2TableImage
from img2table.ocr import TesseractOCR

def detect_and_correct_rotation(pil_img):
    osd = pytesseract.image_to_osd(pil_img)
    angle = 0
    for line in osd.split('\n'):
        if 'Rotate:' in line:
            angle = int(line.split(':')[-1].strip())
            break
    if angle != 0:
        pil_img = pil_img.rotate(-angle, expand=True)
    return pil_img

def extract_tables_from_pdf(pdf_path, ocr):
    tables = []
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            pil_img = page.to_image(resolution=300).original.convert("RGB")
            corrected_img = detect_and_correct_rotation(pil_img)
            # Save to temp file or BytesIO if needed
            temp_path = f"temp_page_{i}.png"
            corrected_img.save(temp_path)
            img2table_img = Img2TableImage(src=temp_path)
            page_tables = img2table_img.extract_tables(ocr=ocr)
            tables.extend(page_tables)
    return tables

# Usage
ocr = TesseractOCR(lang="eng")
tables = extract_tables_from_pdf("RA_2022_30_Oct-187-193.pdf", ocr)
# Now you can process tables as needed, or save them to Excel
```

Detects primary key in the table, eg: states in  staewise accidents data

```python
def detect_row_id_column(df, threshold=0.8):
    for col in df.columns:
        series = df[col].dropna().astype(str)
        non_numeric = series.apply(lambda x: not str(x).replace('.', '', 1).isdigit())
        ratio = non_numeric.mean()
        if ratio >= threshold:
            return col
    # fallback
    return df.columns[0]
```

Detecting Multirow Headers

```python
import pandas as pd

def combine_multirow_headers_clean(df, num_header_rows=2, sep="_", default=""):
    """
    Combine multiple header rows into a single header row.
    - If all header rows for a column are the same non-empty value, use that value.
    - If all are empty/None, use `default`.
    - If all are the same after stripping, use that value.
    - If all are different but non-empty, join unique non-empty values with sep.
    - Ensures resulting columns are unique by appending suffixes if needed.
    """
    headers = df.iloc[:num_header_rows].astype(str).fillna("").applymap(str.strip)
    combined_headers = []
    for c in range(headers.shape[1]):
        col_vals = [v for v in headers.iloc[:, c] if v and v.lower() != "none"]
        unique_vals = list(dict.fromkeys(col_vals))  # preserve order, remove duplicates
        if not unique_vals:
            combined_headers.append(default)
        elif len(unique_vals) == 1:
            combined_headers.append(unique_vals[0])
        else:
            combined_headers.append(sep.join(unique_vals))
    # Ensure column names are unique
    seen = {}
    final_headers = []
    for h in combined_headers:
        if h not in seen:
            final_headers.append(h)
            seen[h] = 1
        else:
            # Append a suffix to make it unique
            final_headers.append(f"{h}_{seen[h]}")
            seen[h] += 1
    df_clean = df.iloc[num_header_rows:].copy().reset_index(drop=True)
    df_clean.columns = final_headers
    return df_clean

```

Function to identify the number of header rows in the table

```python
def guess_header_rows(df, max_rows=10, numeric_threshold=0.2):
    """
    Guess number of header rows by identifying where numeric/tabular data starts.
    """
    for i in range(1, min(max_rows, len(df))):
        row = df.iloc[i]
        num_like = row.apply(lambda x: isinstance(x, (int, float)) or str(x).strip().replace(',', '').replace('.', '').isdigit())
        if num_like.sum() / len(row) >= numeric_threshold:
            return i
    return 1  # fallback
```

```python
# prompt: ADD STEPS TO SKIP TABLES  WITH CONTENT EMPTY MORE THAN 90%, write one individual function

import pandas as pd
def should_skip_table(df, empty_threshold=0.9):
    """
    Checks if a DataFrame should be skipped based on the percentage of empty cells.

    Args:
        df (pd.DataFrame): The DataFrame to check.
        empty_threshold (float): The threshold for the empty cell percentage (0.0 to 1.0).
                                 If empty cell percentage is >= threshold, the table is skipped.

    Returns:
        bool: True if the table should be skipped, False otherwise.
    """
    if df.empty:
        return True  # Skip empty DataFrames
    total_cells = df.size
    if total_cells == 0:
        return True # Avoid division by zero if somehow size is 0

    # Count empty cells (including None, NaN, and empty strings)
    empty_cells = df.isnull().sum().sum() + (df == '').sum().sum()

    empty_ratio = empty_cells / total_cells

    return empty_ratio >= empty_threshold
```
